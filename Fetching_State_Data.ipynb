# Instructions for fetching state data

## Location data
Files that have lat/lon, API Numbers and maybe some basic dates and other status

Because these files can be large, we store them on external drives.

### Alabama
[Start here](https://www.gsa.state.al.us/ogb/gisdata)
- Click, under "Wells": "Wells (Surface Locations...)"
- Move downloaded file to state_raw

### Alaska
[Jump off site](https://dog-soa-dnr.opendata.arcgis.com/)
- Click on "AOGCC Well Surface Location"
Follow directions for New Mexico.

### Arkansas
Follow the instructions [on this page](https://www.aogc.state.ar.us/maps/geostore.aspx)

### California
[Jump off site](https://maps.conservation.ca.gov/oilgas/)
- Lower on page, click "Download" of All Oil and Gas wells.
- Click the SHP icon in the All Wells block.
- Click "View all group content."
- In the "Oil and Gas Wells Gis" shape file, click the ICON. [Direct link?](https://gis.conservation.ca.gov/portal/home/item.html?id=335e036c6a4f4cc39148ca2a9e0389c7)
- Click "Download"  Move file to state_raw

### Colorado
[jump page](https://cogcc.state.co.us/data2.html#/downloads)
- Click on Well surface location files (under GIS Shape files)
- Click on the Well Spots link.  File will be downloaded.  Move to state_raw/colorado as is.

[Guidance page](https://cogcc.state.co.us/documents/data/downloads/COGCC_Download_Guidance.pdf)

### Louisiana
[Top page](https://www.sonris.com/)  Go to GIS Access
- Under Table of Contents, expand "Oil/Gas"
- Where it says "Oil and Gas Wells" click on the little magnifying glass.  That opens up a set of options below.
- Click on the far right icon, Extract Features.  That should start a download of the shapefile, > 50MB.
- Move that file to the state_raw/louisiana directory.

### Mississippi
[Start Here](https://www.ogb.state.ms.us/downloads.php)
- Click on "WellInfoLists"
- Click on the most recent and copy downloaded file to state_raw.  Rename to wells.csv.


### Montana
[Jump off site](http://www.bogc.dnrc.mt.gov/webapps/dataminer/) Go to Well lat/lon dropdown to:
[search page](http://www.bogc.dnrc.mt.gov/webapps/dataminer/Wells/WellSurfaceLongLat.aspx)
set dropdowns to API #, Begins with "25"  search will give 224 pages.
Clicking on the "Text" button downloads the whole set to a tab-delimited .txt file.  Save it in the repository as  "montana/location.csv"

### North Dakota
[Jump off site](https://gis.dmr.nd.gov/)
- Click on GIS Data Download
- Click on OGD_Wells.zip.  That will download it.
- Move it to the state_raw/north_dakota directory

### Ohio
[Jump off site](https://ohiodnr.gov/discover-and-learn/safety-conservation/about-odnr/oil-gas/oil-gas-resources/well-locator). When you click on the "LAUNCH" button, you will get a popup.
- Click on the "Data" tab.
- Click on the "Wells and Wellbores".  A ~20Mb zip file will be automatically downloaded. Put it into the state_raw/Ohio directory.

### Oklahoma
[Jump off site](https://gisdata-occokc.opendata.arcgis.com/)
- Click on RBDMS Wells.
See New Mexico instructions.
**Looks like Oklahoma state data has about 1500 wells that match FF api10, but don't have lat/lon**

### Pennsylvania
[Search page](https://www.depgreenport.state.pa.us/ReportExtracts/OG/OilGasWellInventoryReport)
- Just clicking "export" will download the whole set as a csv.  
- Move to the state_raw/pennsylvania/ folder.

### Texas
[Jump off site](https://www.rrc.texas.gov/resource-center/research/data-sets-available-for-download/)
You want to get to the "Well Layers by County - ArcView Shape file link" [This might be it](https://mft.rrc.texas.gov/link/4e9023eb-e4ee-45b8-81b7-aec1494c1e8e) but better to manually navigate to it.
- Go to the bottom of the page and select 1000 row view (there are more than 250)
- Click on check box at top for ALL rows
- Click the download button; takes at least a minute, over 150 MB
- Move that file to the state_raw directory /texas/  (remove older file).

### Utah
[Jump off site](https://gis.utah.gov/data/energy/oil-gas/)
- Click "Utah Oil and Gas Well Locations: Shapefile" to download, move to state_raw

### Virginia
[Start Here](https://www.energy.virginia.gov/dgoinquiry/frmmain.aspx)
- Towards the bottom of the page, click "Well location download"  (Data are in a different projection and have
to be converted...)


### West Virginia
[Download site](https://data-wvdep.opendata.arcgis.com/datasets/WVDEP::all-dep-wells/explore?location=18.115681%2C-81.613450%2C4.61)
Same as New Mexico.

### Wyoming
[Jump off site](http://pipeline.wyo.gov/rnewuserspage.html)
- in the row "StateWide Well Download", click on "All Wells".  Move file to the state_raw dir and **rename it to Wells.zip**
